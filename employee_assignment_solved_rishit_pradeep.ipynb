```python
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
data['OverTime'] = data['OverTime'].map({'No':0,'Yes':1})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
data['Over18'] = data['Over18'].map({'N':0,'Y':1})
```

```python
from sklearn.preprocessing import LabelEncoder
col = ['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
enc = {}
for column in col:
    enc[column] = LabelEncoder()
    data[column] = enc[column].fit_transform(data[column])
```

```python
data.head()
data.isnull().sum() 
```

```python
x = data.drop(['Attrition'],axis=1)
y = data['Attrition'].values
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state=42)
x_over,y_over = rus.fit_resample(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_over,y_over,test_size=0.2,random_state=42)
 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
dt_predictions = dt_model.predict(x_test)
```

```python
prediction = reg.predict(x_test)
```

```python
lr_accuracy = accuracy_score(y_test, prediction)
lr_precision = precision_score(y_test, prediction)
lr_recall = recall_score(y_test, prediction)
lr_f1 = f1_score(y_test, prediction)
lr_roc_auc = roc_auc_score(y_test, prediction)

print("Logistic Regression Metrics:")
print(f"Accuracy: {lr_accuracy}")
print(f"Precision: {lr_precision}")
print(f"Recall: {lr_recall}")
print(f"F1 Score: {lr_f1}")
print(f"ROC AUC Score: {lr_roc_auc}")



dt_accuracy = accuracy_score(y_test, dt_predictions)
dt_precision = precision_score(y_test, dt_predictions)
dt_recall = recall_score(y_test, dt_predictions)
dt_f1 = f1_score(y_test, dt_predictions)
dt_roc_auc = roc_auc_score(y_test, dt_predictions)



print("\nDecision Tree Metrics:")
print(f"Accuracy: {dt_accuracy}")
print(f"Precision: {dt_precision}")
print(f"Recall: {dt_recall}")
print(f"F1 Score: {dt_f1}")
print(f"ROC AUC Score: {dt_roc_auc}")
```

```python
performance_metrics = {
    "Model": ["Logistic Regression", "Decision Tree"],
    "Accuracy": [lr_accuracy, dt_accuracy],
    "Precision": [lr_precision, dt_precision],
    "Recall": [lr_recall, dt_recall],
    "F1 Score": [lr_f1, dt_f1],
    "ROC AUC": [lr_roc_auc, dt_roc_auc]
}

metrics_df = pd.DataFrame(performance_metrics)
print(metrics_df)
```

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_crash = pd.read_csv(r"C:\Users\rishi\Downloads\car_crashes.csv")

```

```python

# Pairplot
sns.pairplot(car_crash)
plt.title("Pairplot of Car Crashes Data")
plt.show()


```

```python
# Histogram
sns.histplot(car_crash["total"], bins=15, kde=True)
plt.title("Histogram of Total Crashes")
plt.xlabel("Total Crashes")
plt.ylabel("Frequency")
plt.show()
```

```python

# Boxplot
sns.boxplot(x="abbrev", y="total", data=car_crash)
plt.title("Boxplot of Total Crashes by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Total Crashes")
plt.xticks(rotation=90)
plt.show()
```

```python

# Scatterplot
sns.scatterplot(x="alcohol", y="total", data=car_crash)
plt.title("Scatterplot of Alcohol vs. Total Crashes")
plt.xlabel("Percentage of Alcohol-Impaired Drivers")
plt.ylabel("Total Crashes")
plt.show()
```
